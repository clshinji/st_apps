```python
# モジュールインポート、初期設定

import tabula
import pandas as pd

column_names_pg1 = [
    '分類',
    '商品名',
    '不要',
    '卵',
    '乳',
    '小麦',
    'えび',
    'かに',
    'そば',
    '落花生',
    'あわび',
    'いか',
    'いくら',
    'オレンジ',
    'キウイ',
    '牛肉',
    'クルミ',
    '鮭',
    'さば',
    '大豆',
    '鶏肉',
    'バナナ',
    '豚肉',
    '松茸',
    '桃',
    '山芋',
    'りんご',
    'ゼラチン',
    'ごま',
    'カシューナッツ',
    'アーモンド',
]

column_names_pg2 = [
    '分類',
    '商品名',
    '卵',
    '乳',
    '小麦',
    'えび',
    'かに',
    'そば',
    '落花生',
    'あわび',
    'いか',
    'いくら',
    'オレンジ',
    'キウイ',
    '牛肉',
    'クルミ',
    '鮭',
    'さば',
    '大豆',
    '鶏肉',
    'バナナ',
    '豚肉',
    '松茸',
    '桃',
    '山芋',
    'りんご',
    'ゼラチン',
    'ごま',
    'カシューナッツ',
    'アーモンド',
]

print(f'column counts(p.1):{len(column_names_pg1)}')
print(f'column names (p.1):{column_names_pg1}')

# PDFファイルから表を読み込む
df = tabula.read_pdf('allergy.pdf', pages='all')
```

```python
# 読み込まれたデータの確認
print(f'data type(p.1): {type(df[0])}')

# ページ数を取得する
page_counts = len(df)
print(f'page counts:{page_counts}')

# 全ページのカラム数が一致していることを確認する
# p.1のみ31列として読み込まれることが多い？
for i in range(len(df)):
    df_org = df[i].copy()
    # print(f'column names :{df_org.columns}')
    print(f'page No.:{i}, column counts:{len(df_org.columns)}')

# df check -> p.1のデータフレームだけ表示してみる
df[0].head()
```

```python
# 1ページ目だけ先に読み込む
# 1ページ目の表を取得する
df_org = df[0].copy()
# カラム名を設定する
df_org.columns = column_names_pg1

# 「'●', '○'」を含まない行を削除する
df_allergy = df_org[df_org[column_names_pg1].isin(['●', '○']).any(axis=1)].copy()

# 1ページ目だけ31ページある場合に実行
# 1ページ目の不要な列を削除する
df_allergy = df_allergy.drop(columns='不要')


# 2ページ目以降の表を取得して結合する
for i in range(1, len(df)):
    df_org = df[i].copy()
    df_org.columns = column_names_pg2
    # print(f'column names :{df_org.columns}')
    print(f'page No.:{i}, column counts:{len(df_org.columns)}')
    # 「'●', '○'」を含まない行を削除する
    df_org = df_org[df_org[column_names_pg2].isin(['●', '○']).any(axis=1)].copy()
    # df_allergyにdf_orgを縦方向に連結する
    df_allergy = pd.concat([df_allergy, df_org], axis=0, ignore_index=True).copy()
    print(f' -> df_allergy shape: {df_allergy.shape}')

df_allergy
```

```python
# 作成したdf_allergyをcsv形式で保存する
df_allergy.to_csv('allergy_table.csv', encoding='shift-jis')
```
